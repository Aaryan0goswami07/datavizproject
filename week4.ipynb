```python
# ---------------------------------------------
# Complete Data Analysis & Visualization Project
# Week 4
# ---------------------------------------------

import pandas as pd
import matplotlib.pyplot as plt
import os

# Create folder for saving charts
os.makedirs("visualizations", exist_ok=True)

try:
    # Load dataset using your exact file path
    file_path = r"C:\Users\Aryan\Downloads\sales_data (2).csv"
    df = pd.read_csv(file_path)
    print("âœ… Dataset loaded successfully!\n")

    # Show first 5 rows
    print("First 5 rows of dataset:\n")
    print(df.head(), "\n")

    # Dataset info
    print("Dataset Information:\n")
    df.info()

    # Check missing values
    print("\nMissing Values:\n", df.isnull().sum())

    # Remove duplicate rows
    df.drop_duplicates(inplace=True)

    # ------------------ DATA ANALYSIS ------------------

    # Try to find numeric columns automatically
    numeric_cols = df.select_dtypes(include='number').columns

    print("\nStatistical Summary:\n")
    print(df[numeric_cols].describe())

    # ------------------ VISUALIZATION 1 ------------------
    # Bar Chart for most frequent category column

    category_col = df.select_dtypes(include='object').columns[0]
    df[category_col].value_counts().plot(kind='bar')
    plt.title(f"Distribution of {category_col}")
    plt.xlabel(category_col)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.savefig("visualizations/bar_chart.png")
    plt.show()

    # ------------------ VISUALIZATION 2 ------------------
    # Pie Chart for second categorical column (if exists)

    cat_cols = df.select_dtypes(include='object').columns

    if len(cat_cols) > 1:
        df[cat_cols[1]].value_counts().plot(kind='pie', autopct='%1.1f%%')
        plt.title(f"{cat_cols[1]} Distribution")
        plt.ylabel("")
        plt.tight_layout()
        plt.savefig("visualizations/pie_chart.png")
        plt.show()

    print("\nðŸ“Š Analysis and visualization completed successfully!")

except FileNotFoundError:
    print("âŒ Dataset file not found. Please check the path.")

except Exception as e:
    print("âš ï¸ An error occurred:", e)
```
